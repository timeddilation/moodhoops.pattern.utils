```python
import moodhoops as mh
from moodhoops.utils.image_io import preview_pattern
```

# Pattern Slow Down

Within a single mode, only one speed setting can be set.
In order to simulate specific patterns in the mode being on slower speeds, we can stretch the image vertically.

By knowing the current speed setting of the mode, and the desired speed setting for the single pattern, we can create a modified pattern that will appear as if it is on the lower speed setting, while using the mode's higher speed setting.

```python
foo = mh.load_image("tests/bmps/test_stretch.bmp")
preview_pattern(foo)
```

```python
bar = mh.slow_down_pattern(
    canvas=foo,
    mode_speed=300,
    desired_speed=50,
)
preview_pattern(bar)
```

```python
# save image to exports
mh.save_image(bar, "test_stretch_slowed_down.bmp")
```

# Swap Colors

You may have a pattern with a set of colors, and you want to find and replace a single color or a set of colors.

By definig a list of **from** and **to** colors, we can progamatically replace or even swap colors.

```python
from moodhoops.features.swap_colors import swap_colors

foo = mh.load_image("tests/bmps/rgb.bmp")
preview_pattern(foo)
```

```python
# find and replace one color
bar = swap_colors(foo, color_maps=[
    {"from": "#00FF00", "to": "#FFFFFF"},
])
preview_pattern(bar)
```

```python
# swap two colors with each other
bar = swap_colors(foo, color_maps=[
    {"from": "#FF0000", "to": "#0000FF"},
    {"from": "#0000FF", "to": "#FF0000"},
])
preview_pattern(bar)
```
